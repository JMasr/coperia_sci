```python
import pandas as pd
from pathlib import Path
```

```python
ROOT_PATH = Path(r'C:\Users\jmram\OneDrive\Documents\GitHub\coperia_tdd\data')
coperia_df = pd.read_csv(ROOT_PATH / 'coperia_metadata.csv')
coperia_df
```

```python

# Change the NaN values to False
coperia_df['covid'] = False

# Cast column duration to float
for i in coperia_df.index:
    duration_value = coperia_df.at[i, 'duration']
    if isinstance(duration_value, str): 
        float_number_as_string_with_comma = duration_value
        float_number_as_string_with_dot = float_number_as_string_with_comma.replace(',', '.')
        coperia_df.at[i, 'duration'] = float(float_number_as_string_with_dot)
    
    if duration_value == 0.0:
        print(f"Data Inconsistent at Index: {i} - {coperia_df.at[i, 'audio_id']} - {duration_value}")
        

coperia_df['age'] = coperia_df['age'].astype(int)
coperia_df['covid'] = coperia_df['covid'].astype(bool)
coperia_df['duration'] = coperia_df['duration'].astype(float)
coperia_df['long_covid'] = coperia_df['long_covid'].astype(bool)

# Change the value of the column duration using the audio_id column value
def change_a_vaule_usign_audio_id(df, audio_i, column_to_edit, value_to_change):
    original_value = df.loc[df['audio_id'] == audio_i, column_to_edit].values[0]
    if original_value == value_to_change:
        print(f"Value already changed to {value_to_change}")
    else:
        df.loc[df['audio_id'] == audio_i, column_to_edit] = value_to_change        
        
change_a_vaule_usign_audio_id(coperia_df,'7a050268-0d63-4a99-88ad-fcb86a19f976', 'duration', 0.90)
change_a_vaule_usign_audio_id(coperia_df,'832644f4-202d-4908-bc57-8631e8003c3f', 'duration', 0.90)
coperia_df.drop_duplicates(subset=['audio_id'], keep='first', inplace=True)

coperia_df
```

```python
coperia_df.to_csv(ROOT_PATH / 'coperia_metadata.csv', index=False)
coperia_df.to_csv(ROOT_PATH / 'coperia_metadata_v2_for_xls.csv', index=False, decimal=',', sep=';')
```
